```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
```

```python
cityB_DF = pd.read_csv("hiroshima_challengedata.csv")
cityC_DF = pd.read_csv("sapporo_challengedata.csv")
cityD_DF = pd.read_csv("kumamoto_challengedata.csv")
```

```python
cityD_test = cityD_DF[cityD_DF['x']==999].drop(['x', 'y'],axis=1)
cityD_test.head()
```

```python
cityD_train = cityD_DF[cityD_DF['uid'].isin(cityD_test['uid'].unique())]
cityD_train = cityD_train[cityD_train['x']!=999]
cityD_train.head()
```

```python
# define model
model = KNeighborsRegressor()

x_list = []
y_list = []

for i in cityD_test['uid'].unique():
    # fit model
    model.fit(cityD_train[cityD_train['uid']==i][['t']], cityD_train[cityD_train['uid']==i][['x', 'y']])
    result = model.predict(cityD_test[cityD_test['uid']==i][['t']])
    for res in result:
        x_list.append(res[0])
        y_list.append(res[1])
    print(f"i={i} done")

cityD_test['x'] = x_list
cityD_test['y'] = y_list
cityD_test.head()
```
